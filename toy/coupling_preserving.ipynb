### SBALIGN

```python
Experiment.run(f"--method=sbalign --dataset=X --h_dim=64  --n_layers=2  --n_epochs=20  --reg_weight=1.  --timestep_emb_dim=32  --in_dim=2 --out_dim=2 --max_diffusivity=0.5  --H=0.5  --K=0 --norm=True --use_drift_in_doobs=True  --activation=silu").save("x_vis_sbalign")

sbalign_coupling = Experiment.load("x_vis_sbalign")
plt.axis("off")
plot_marginals(sbalign_coupling.get_marginals(samples_num=1000))
plt.show()
plt.close()

sbalign_samples = sbalign_coupling.sample(samples_num=100, trials_num=1)
plot_trajectories(sbalign_samples,title='SBALIGN')
```
