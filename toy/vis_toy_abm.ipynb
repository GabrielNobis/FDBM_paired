# ABM on Moons

### Training

```python
%load_ext autoreload

%autoreload 2

import warnings
# Suppress non-critical warnings (including verbose local file paths). Uncomment to display warnings.
warnings.filterwarnings("ignore")

import os
os.chdir("..")

from toy.experiment import *

Experiment.run(f"--method=abm --dataset=moon --h_dim=64  --n_layers=2  --n_epochs=20  --reg_weight=1.  --timestep_emb_dim=32  --in_dim=4 --out_dim=2 --max_diffusivity=0.6  --H=0.5  --K=0 --activation=silu").save("moon_vis_abm")
```

### Visual evaluation

```python
moon = Experiment.load("moon_vis_abm")

plt.axis("off")
plt.axis("equal")
plot_marginals(moon.get_marginals())
samples = moon.sample(samples_num=1000, trials_num=1)
plot_multiple_marginals(samples, skip_step=4) 
plt.show()
plt.close()

plt.axis("off")
plt.axis("equal")

plot_matchings(*samples[[0,-1]])
plot_multiple_marginals(samples, skip_step=99, labels=[r"$t_0$", r"$t_1$"], alpha=1)
plt.show()
plt.close()
```

# ABM on T-shaped

### Training

```python
Experiment.run(f"--method abm --dataset=diagonal_matching_inverse  --h_dim=32  --n_layers=3  --n_epochs=20  --reg_weight=1.  --timestep_emb_dim=32 --in_dim=4 --max_diffusivity=0.2 --H=0.5 --K=0 --activation=selu").save("tshaped_vis_abm")
```

### Visual Evaluation

```python
tshaped = Experiment.load('tshaped_vis_abm')

plt.axis("off")
plt.axis("equal")
plot_marginals(tshaped.get_marginals(), alpha=.2)
samples = tshaped.sample(samples_num=1000, trials_num=1)
plot_multiple_marginals(samples, skip_step=4)
plt.show()
plt.close()

plt.axis("off")
plt.axis("equal")
plot_matchings(*samples[[0,-1]])
plot_multiple_marginals(samples, skip_step=100, labels=[r"$t_0$", r"$t_1$"], alpha=1)
plt.show()
plt.close()
```
